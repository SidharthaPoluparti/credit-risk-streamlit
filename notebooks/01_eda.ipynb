Split features and target

```python
from sklearn.model_selection import train_test_split

X = model_df.drop("SeriousDlqin2yrs", axis=1)
y = model_df["SeriousDlqin2yrs"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report

log_reg = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    n_jobs=-1
)

log_reg.fit(X_train, y_train)
```

```python
train_pred = log_reg.predict_proba(X_train)[:, 1]
test_pred  = log_reg.predict_proba(X_test)[:, 1]

print("Train AUC:", roc_auc_score(y_train, train_pred))
print("Test AUC:", roc_auc_score(y_test, test_pred))
```

```python
from sklearn.metrics import accuracy_score

# default threshold = 0.5
y_test_pred_label = (test_pred >= 0.5).astype(int)

accuracy = accuracy_score(y_test, y_test_pred_label)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import accuracy_score

# Train predictions
y_train_pred_label = (train_pred >= 0.5).astype(int)

# Test predictions
y_test_pred_label = (test_pred >= 0.5).astype(int)
```

```python
train_accuracy = accuracy_score(y_train, y_train_pred_label)
test_accuracy  = accuracy_score(y_test, y_test_pred_label)

print("Train Accuracy:", train_accuracy)
print("Test Accuracy :", test_accuracy)
```

lets calculate VIF to see multicoliniarity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

X_vif = X_train.copy()

vif_df = pd.DataFrame()
vif_df["feature"] = X_vif.columns
vif_df["VIF"] = [
    variance_inflation_factor(X_vif.values, i)
    for i in range(X_vif.shape[1])
]

vif_df.sort_values("VIF", ascending=False)
```

VIF values were checked. Moderate multicollinearity was observed for age and open credit lines (VIF ~6), which is acceptable given their strong business relevance.

```python
decile_df = pd.DataFrame({
    "y_true": y_test,
    "score": test_pred
})

decile_df["decile"] = pd.qcut(decile_df["score"], 10, labels=False)

decile_summary = (
    decile_df
    .groupby("decile")
    .agg(
        total=("y_true", "count"),
        defaults=("y_true", "sum"),
        default_rate=("y_true", "mean")
    )
    .sort_index(ascending=False)
)

decile_summary
```
